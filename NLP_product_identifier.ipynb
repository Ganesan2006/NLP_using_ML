# NLP

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
from nltk.stem.porter import PorterStemmer
```

```python
import pandas as pd
data=pd.read_csv("data.csv")
```

```python
data.head()
```

```python
text=data["Text"]
```

```python
text.head(1).values
```

```python
import re
def lower(string: str):
    return string.lower()
low_text=[lower(string) for string in text]
```

```python
low_text
```

```python
import re

def remove_spl_num(string: str) -> str:
    return re.sub(r'[^a-zA-Z ]', '', string)
word_without_num_spl=[remove_spl_num(string) for string in low_text]
```

```python
word_without_num_spl[1]
```

```python
import nltk
nltk.download("punkt_tab")
from nltk.tokenize import word_tokenize
tokenized=[word_tokenize(word) for word in word_without_num_spl]
```

```python
tokenized[0]
```

```python
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords

def remove_stopwords(paragraph: str) -> str:
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in paragraph if word not in stop_words]
    return ' '.join(filtered_words)


result = [remove_stopwords(word) for word in tokenized]
print(result)
```

```python
porter=PorterStemmer()
stemed_word=[porter.stem(word) for word in result]
```

```python
stemed_word[0]
```

```python
x=np.array(stemed_word)
y=data["Cat1"]
```

```python
print(x.shape)
```

```python
print(y.shape)
```

```python
vector=TfidfVectorizer()
tfidf=vector.fit_transform(x)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(tfidf,y,test_size=0.3,random_state=0)
model=RandomForestClassifier(max_depth=50,random_state=0)
model.fit(xtrain,ytrain)
```

```python
predictt=model.predict(xtest)
accuracy_score(predictt,ytest)
```
